```python
housing_data["sqft_living"] = housing_data["sqft_living"].fillna(housing_data["sqft_living"].median())
housing_data["sqft_lot"] = housing_data["sqft_lot"].fillna(housing_data["sqft_lot"].median())
housing_data["yr_built"]=housing_data["yr_built"].fillna(housing_data["yr_built"].median())
```

FINDING AND RECTIFING OUTLIERS

```python
sns.boxplot(housing_data['price'])
plt.title("Distribution of Price")
plt.show()
```

```python
# Calculate Q1, Q3, and IQR
Q1 = housing_data['price'].quantile(0.25)
Q3 = housing_data['price'].quantile(0.75)
IQR = Q3 - Q1

# Define bounhousing_data
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Calculate median
median_price = housing_data['price'].median()

# Replace outliers with median
housing_data['price'] = housing_data['price'].apply(lambda x: median_price if x < lower_bound or x > upper_bound else x)
```

FEATURE ENGINEERING

```python
import datetime
current_year = datetime.datetime.now().year
housing_data['property_age'] = current_year - housing_data['yr_built']
```

```python
housing_data['total_sqft'] = housing_data['sqft_living'] + housing_data['sqft_basement']
housing_data['price_per_sqft'] = housing_data['price'] / housing_data['total_sqft']
```

TEST STATISTICS

```python
from scipy.stats import ttest_ind

# Group prices
waterfront = housing_data[housing_data['waterfront'] == 1]['price']
non_waterfront = housing_data[housing_data['waterfront'] == 0]['price']

t_stat, p_val = ttest_ind(waterfront, non_waterfront, equal_var=False)

# Print results
print("T-Test: Price Comparison  Waterfront vs. Non-Waterfront")
print(f"T-statistic : {t_stat:.2f}")
print(f"P-value     : {p_val:.4e}")

# Hypothesis decision
if p_val < 0.05:
    print("Result      : Reject the null hypothesis (significant difference in price).")
else:
    print("Result      : Fail to reject the null hypothesis (no significant difference in price).")
```
